```python
# Load  dataset 
df = pd.read_csv('BADR Exel Sheet.csv')

#Convert Date Columns to DateTime:
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df['EndDate'] = pd.to_datetime(df['EndDate'], errors='coerce')

# Check for missing values in the dataset
missing_values = df.isnull().sum()
print(missing_values)
```

2. Encode Categorical Variables

```python
# Drop the 'ProjectID' and 'ProjectName' columns
df.drop(columns=['ProjectID', 'ProjectName'], inplace=True)

# Define columns with missing values
columns_with_missing_values = ['HourlyRate', 'OvertimeRate', 'ProductivityRate', 
                               'MarketWageRate', 'FringeBenefits', 'ContingencyCost']

# Create an imputer object with a mean filling strategy
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to the columns with missing values
df[columns_with_missing_values] = imputer.fit_transform(df[columns_with_missing_values])

# Check to ensure there are no missing values left
print(df.isnull().sum())
```

```python


# One-hot encode categorical variables 
categorical_columns = [ 'ProjectType', 'ProjectComplexity', 'LaborDemand','WeatherImpact']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

df.head()
```

3. Feature Engineering

```python
# Create new feature: Project Duration
df['ProjectDuration'] = (df['EndDate'] - df['StartDate']).dt.days

print(df[['StartDate', 'EndDate', 'ProjectDuration']].head())
```

4. Normalize/Scale Numerical Features:

```python
from sklearn.preprocessing import StandardScaler

# List of numerical features to scale
numerical_features = ['ProjectSize', 'NumWorkers', 'HourlyRate', 'OvertimeRate', 'ProductivityRate',
                      'MarketWageRate', 'InflationRate', 'MinWage', 'FringeBenefits', 'IndirectCosts',
                      'EquipmentCost', 'MaterialCost', 'ContingencyCost', 'ProjectDuration']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the numerical features
df[numerical_features] = scaler.fit_transform(df[numerical_features])

df[numerical_features].head()
```
